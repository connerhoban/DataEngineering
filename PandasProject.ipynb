```python
#Conner Hoban
#July 8



#1
import pandas as pd
df1 = pd.read_csv(r'zomato.csv', encoding = "ISO-8859-1")
df2 = pd.read_excel('Country-Code.xlsx')
df = df1.groupby('Restaurant Name')['Votes'].max()
df = pd.DataFrame(df,columns=['Votes'])
df.sort_values('Votes', ascending = False)
df1.info()
```

```python
#2
df = df1[['City','Restaurant Name','Rating text']]
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
#df.set_index('City', inplace = True)
filt = (df['Rating text'].str.contains('Poor|Not rated|Very Good', na = False))
df_rating = df[filt]
df_tabs = pd.crosstab(df_rating['City'],df_rating['Rating text'])
df_tabs
```

```python
#3
pd.set_option('display.max_rows', 100)
frames = pd.merge(df1,df2,on = 'cc')
frames = frames[['Country','City','Average Cost for two']]

g=frames.groupby('Country').agg({'Average Cost for two': 'mean','City': 'max'})
g

    

```

```python
#4
frames = pd.merge(df1,df2,on = 'cc')
mask = frames['Country'] == 'India'
frames[mask]
frames.groupby('Locality')['Restaurant Name'].count()
```

```python
#5
frames = pd.merge(df1,df2,on = 'cc')
g=frames.groupby('Country')

out = {}
for name, df in g:
    out[name] = df.groupby('City')['Restaurant ID'].count().idxmax()
    
out
```

```python
#6
frames = pd.merge(df1,df2,on = 'cc')
g = frames.groupby('Restaurant Name')

out = {}
for name, df in g:
    out[name] = df['Restaurant ID'].count()



out.values()
max_key = max(out, key = out.get)
max_franchise = {max_key: out[max_key]}
max_franchise

    
```

```python
#7
frames = pd.merge(df1,df2,on = 'cc')
mask = frames['Has Online delivery'] == 'Yes'
frames = frames[mask]
frames["Restaurant Name"].count()
```

```python
#8
frames = pd.merge(df1,df2,on = 'cc')
mask = frames['Has Table booking'] == 'Yes'
frames = frames[mask]
frames["Restaurant Name"].count()
```

```python
#9
frames = pd.merge(df1,df2,on = 'cc')
g=frames.groupby('City')

out = {}
for name, df in g:
    out[name] = df['Restaurant ID'].count()
    
max_key = max(out, key = out.get)
max_restaurants = {max_key: out[max_key]}
max_restaurants

```
