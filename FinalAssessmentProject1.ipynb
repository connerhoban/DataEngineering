```python
df = pd.read_csv('cars.csv')
df.head(10)
```

```python
# look for duplicate entries
print(df.shape[0])
df = df.drop_duplicates()
print(df.shape[0])
```

```python
df.info()
```

```python
#1 What percentage of cars will be suitable for a family (i.e. num of doors=4, price <20,000 & mileage >17)?

# cleansing the price column and casting as an integer, replacing '?' with mean value of column
df['price'].replace(to_replace = '?', value = 0, inplace = True)
df['price'] = df['price'].astype(int)
df['price'].replace(to_replace = 0, value = np.mean(df['price']), inplace = True)


filt = ((df['num of doors'].str.contains('four'))&(df['price']<20000)&(df['city mpg']>17)&(df['highway mpg']>17))
#df[filt]
round(df[filt].shape[0]/df.shape[0],2)
```

```python
#2 Which company has generated more car options for customers?

df1 = df.groupby('make')['make'].count()
df1.sort_values(ascending = False).head(3)
```

```python
#3 What is the ratio of diesel cars to that gas ones?

df1 = df.groupby('fuel-type')['fuel-type'].count()
print(df1)
print('\nThe ratio of diesel cars to gas cars is:', round(20/185,2))
```

```python
#4 What is the count of performance cars present in the dataset (horsepower > 150)?

# cleanse the horsepower column and replace ? with mean value, cast column as type int
df['horsepower'].replace(to_replace = '?', value = 0, inplace = True)
df['horsepower'] = df['horsepower'].astype(int)
df['horsepower'].replace(to_replace = 0, value = np.mean(df['horsepower']), inplace = True)

df[df['horsepower']>150].shape[0]
```

```python
#5 Which is the most compact among all cars?

pd.set_option('display.max_columns', 30)
df.sort_values(by = ['width','height','length'], ascending = True).head(10)
```

```python
#6 What are the main factors that are associated with the mileage of a car?

print(df.corr()[['city mpg','highway mpg']],'\n')

# factors with strongest correlation to city and highway mpg are horsepower, curb weight, and length
# additional factors with slightly less correlation are width and engine size
# categorical variables that affect mileage of a car are fuel type and number of cylinders


# compare fuel types
df1 = df.groupby('fuel-type')[['city mpg','highway mpg']].mean()
print(df1.sort_values(['city mpg','highway mpg'],ascending = False),'\n')

# compare number of cylinders
df2 = df.groupby('num of cylinders')[['city mpg','highway mpg']].mean()
print(df2.sort_values(['city mpg','highway mpg'],ascending = False),'\n')

# compare body style
df3 = df.groupby('body style')[['city mpg','highway mpg']].mean()
print(df3.sort_values(['city mpg','highway mpg'],ascending = False),'\n')

# compare drive wheels
df4 = df.groupby('drive wheels')[['city mpg','highway mpg']].mean()
print(df4.sort_values(['city mpg','highway mpg'],ascending = False),'\n')

# compare engine type
df5 = df.groupby('engine type')[['city mpg','highway mpg']].mean()
print(df5.sort_values(['city mpg','highway mpg'],ascending = False),'\n')

# compare fuel system
df6 = df.groupby('fuel system')[['city mpg','highway mpg']].mean()
print(df6.sort_values(['city mpg','highway mpg'],ascending = False),'\n')






```

```python
#7 What percentage of cars are budget-friendly (price < 10,000)?

round(df[df['price']<10000].shape[0]/df.shape[0],2)
```

```python
#8 Which cars are the most efficient of all (city mpg >= 30)?

df1 = df[df['city mpg'] >= 30]
df1.sort_values(by = ['city mpg'], ascending = False)
```

```python
#10 Which feature of the car affects the most to the pricing?

print(df.corr()[['city mpg','highway mpg']],'\n')

# feature with most correlation to price is engine size, so it affects pricing the most
# the categorical variables that affect price are number of cylinders, fuel types, and drive wheels

# compare fuel types
df1 = df.groupby('fuel-type')['price'].mean()
print(df1.sort_values(ascending = False),'\n')

# compare number of cylinders
df2 = df.groupby('num of cylinders')['price'].mean()
print(df2.sort_values(ascending = False),'\n')

# compare body style
df3 = df.groupby('body style')['price'].mean()
print(df3.sort_values(ascending = False),'\n')

# compare drive wheels
df4 = df.groupby('drive wheels')['price'].mean()
print(df4.sort_values(ascending = False),'\n')

# compare engine type
df5 = df.groupby('engine type')['price'].mean()
print(df5.sort_values(ascending = False),'\n')

# compare fuel system
df6 = df.groupby('fuel system')['price'].mean()
print(df6.sort_values(ascending = False),'\n')


```
